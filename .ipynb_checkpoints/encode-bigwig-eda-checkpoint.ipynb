```python
# useful libraries to import

import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn import manifold

import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
bp_formatter = EngFormatter('b') 
# nice way to format ticks as human-readable: ax.xaxis.set_major_formatter(bp_formatter)

from hwutils import *
```

```python
# load dataFrame of bigWigs from ENCODE (encodeproject.org/), binned to 10kb resolution across chromosome 10.
# note that the first three columns are chrom,start,end and the other columns are labeled by bigWig file accession.
df = pd.read_table('./data/ENCODE_GRCh38_binned_subset.tsv')

# load metadata from ENCODE for bigwig files. 
# can be queried as follows: bigwig_metadata.query("`File accession`==@ df_column_name ")
bigwig_metadata = pd.read_table('./data/ENCODE_GRCh38_bigWig_metadata.tsv')
```

```python
final_metadata = bigwig_metadata[bigwig_metadata['File accession'].isin(df.columns[3:])]
```

```python
final_metadata = final_metadata.set_index('File accession').reindex(df.columns[3:]).reset_index()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(df['start'], df['ENCFF230YBR'], label='Line 1',color='red')
axs[0, 0].set_title('ENCFF230YBR')
axs[0, 1].plot(df['start'], df['ENCFF953CVO'], label='Line 2',color='green')
axs[0, 1].set_title('ENCFF953CVO')
axs[1, 0].plot(df['start'], df['ENCFF520IHF'], label='Line 3',color='blue')
axs[1, 0].set_title('ENCFF520IHF')
axs[1, 1].plot(df['start'], df['ENCFF730SPR'], label='Line 4',color='purple')
axs[1, 1].set_title('ENCFF730SPR')
```

- Use scikit-learn to perform PCA, and make a scatterplot of PC1 vs PC2.


```python
X_no_standardized = np.transpose(df.iloc[:,3:].values)
pca = PCA(n_components=2)
principalComponents_no_standardized = pca.fit_transform(X_no_standardized)

principalDf_no_standardized = pd.DataFrame(data = principalComponents_no_standardized
             , columns = ['principal component 1', 'principal component 2'])
plt.scatter(principalDf_no_standardized['principal component 1'], principalDf_no_standardized['principal component 2'], label='Scatter Plot', color='red', marker='o')

# Add labels and a title
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA')
```

- Try to use the experiment metadata to understand and remove outliers. Try labeling or coloring points by various metadata columns. Were any columns in the metadata useful for outlier removal? Note that `sklearn.preprocessing.LabelEncoder()` can be useful for transforming text strings to categories, and `plt.text` can be used to overlay labels with points.

```python
import math

# Determine the number of subplots based on the columns
num_plots = len(final_metadata.columns)
num_cols = 3  # for example, 3 columns of subplots
num_rows = math.ceil(num_plots / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.ravel()  # Convert the 2D array of axes into a flat 1D array

for idx, column in enumerate(final_metadata.columns):
    # Convert column values to labels
    le = LabelEncoder()
    labels = le.fit_transform(final_metadata[column].fillna('NA'))  # Fill NaN values with 'NA' for encoding

    scatter = axes[idx].scatter(principalDf_no_standardized['principal component 1'], principalDf_no_standardized['principal component 2'], c=labels, edgecolor='k', cmap='jet', alpha=0.6)
    fig.colorbar(scatter, ax=axes[idx], orientation='vertical')
    axes[idx].set_xlabel('Principal Component 1')
    axes[idx].set_ylabel('Principal Component 2')
    axes[idx].set_title(f'PCA with {column} Labels')
    axes[idx].grid(True)

# Hide any remaining unused subplots
for ax in axes[num_plots:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

- Which Assays or Experiment Targets show broad vs narrow patterns? Is this consistent across cell types? Does this relate to the patterns seen in PCA? One way to investigate the characteristic scale is by computing the autocorrelation.


```python
# Compute auto-correlation for each row
auto_corr_by_row = []

for row in X_no_standardized:
    # Compute the auto-correlation for the row
    auto_corr = np.correlate(row, row, mode='full')
    auto_corr = auto_corr[len(row)-1:]  # Take only the positive lags
    auto_corr /= np.max(auto_corr)  # Normalize the auto-correlation
    auto_corr_by_row.append(auto_corr)

# Convert the list of auto-correlations to a NumPy array
auto_corr_by_row = np.array(auto_corr_by_row)

#make pandas dataframe
auto_correlation_dataframe = pd.DataFrame(auto_corr_by_row )
auto_correlation_dataframe['Experiment target']=final_metadata['Experiment target']
auto_correlation_dataframe['Biosample term name']=final_metadata['Biosample term name']

mean_by_category = auto_correlation_dataframe.groupby('Experiment target').mean()

#set threshold, if >0.9 then broad pattern, if <0.9 narrow pattern
broad = mean_by_category[mean_by_category.iloc[:,1]>0.9].index.tolist()

#display protein target that have broad pattern
broad
```

I computed the auto correlation for each row (accession number) and aggregated and calculate mean for each lag for each experiment target. I set a arbitary threshold 0.9 if the second lag is bigger than 0.9 that means it is a broad pattern and if it is smaller than 0.9 that means the target protein is a narrow pattern. The above shows the experiment target that have the broad pattern. And the below shows the experiment target that have a narrow pattern. 

```python
narrow = mean_by_category[mean_by_category.iloc[:,1]<=0.9].index.tolist()
narrow
```

To show if pattern is consistant across cell types, I take CTCF-human as an example.

```python
auto_correlation_dataframe[auto_correlation_dataframe['Experiment target'] == 'CTCF-human']
```

It looks like the pattern which is narrow here is consistant across different cell types denoted in the last column.

```python
principalDf_no_standardized['Experiment target'] = final_metadata['Experiment target']
principalDf_no_standardized['pattern'] = 'narrow'
principalDf_no_standardized.loc[principalDf_no_standardized['Experiment target'].isin(broad),'pattern'] = 'broad'
import seaborn as sns
sns.scatterplot(data=principalDf_no_standardized, x='principal component 1', y='principal component 2', hue='pattern', palette='Set1')
```

The pca actually split broad vs narrow pattern proteins!

- How many PCs are needed to explain 90% of the variance in the data? Illustrate this with a scree plot (https://en.wikipedia.org/wiki/Scree_plot). 


```python
pca = PCA()
pca.fit(X_standardized)
# Extract the explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_
# Calculate the cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
# Plot the scree plot
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, label='Explained Variance Ratio', align='center')
plt.step(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, where='mid', label='Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Scree Plot for PCA')
plt.legend()
plt.show()
```

```python
cumulative_explained_variance[45]
```

It looks like 45 PCs are needed to explain 90% of the variance in the data

- How different is the dimensionality reduction into two dimensions for PCA from that obtained using MDS (multi-dimensional scaling)? What methods could be used to determine the similarity? Illustrate with a plot.


```python
# Calculate dissimilarity matrix (Euclidean distances in this example)
dissimilarity_matrix = np.sqrt(((X_standardized[:, None] - X_standardized) ** 2).sum(axis=2))
```

```python
mds = manifold.MDS(n_components=2, dissimilarity='precomputed')
X_mds = mds.fit_transform(dissimilarity_matrix)

# Plot the MDS representation
plt.figure(figsize=(8, 6))
plt.scatter(X_mds[:, 0], X_mds[:, 1],color='red', marker='o')
plt.xlabel('MDS Dimension 1')
plt.ylabel('MDS Dimension 2')
plt.title('MDS')
plt.show()
```

While PCA tries to maximize variance along each PC, MDS tries to preserve pairwise distances or dissimilarities as much as possible. Similarity can be determined through different ways, for example I used Euclidean distances in this example to denote similarity between two points.
